```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-100,100,1)
y=x**2
x1=np.arange(-100, 100, 10)
y1=x1**2



plt.scatter(x1, y1, linewidth=2, color='blue', marker ='s', s=90, edgecolor='black', zorder=1)
plt.plot(x, y, linewidth=3, linestyle ='--', color='red', zorder=0)

plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')
plt.show()







```
