```python
import matplotlib.pyplot as plt


import numpy as np
x=np.arange(1.1,100,0.5)
y=(x**3-4)/((x-1)**3)
plt.plot(
     x,y,color="g",linewidth=4)


plt.show()
```

```python
import matplotlib.pyplot as plt


import numpy as np
x=np.arange(-100,100,0.5)
x = list(x)
x.remove(1)
x = np.array(x)
y=(x**3-4)/((x-1)**3)
plt.plot(
     x,y,color="g",linewidth=4)


plt.show()
```
